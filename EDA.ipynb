```python
from pykrx import stock
import pandas as pd

# 1. 코스피와 코스닥 티커만 각각 가져오기
kospi_tickers = stock.get_market_ticker_list(market="KOSPI")
kosdaq_tickers = stock.get_market_ticker_list(market="KOSDAQ")

# 2. 리스트 합치기
target_tickers = kospi_tickers + kosdaq_tickers

# 3. 데이터프레임 생성 시 시장 구분(Exchange) 라벨링
rows = []
for t in kospi_tickers:
    rows.append({'ticker': t, 'name': stock.get_market_ticker_name(t), 'exchange': 'KOSPI'})
for t in kosdaq_tickers:
    rows.append({'ticker': t, 'name': stock.get_market_ticker_name(t), 'exchange': 'KOSDAQ'})

df_krx = pd.DataFrame(rows)
```
